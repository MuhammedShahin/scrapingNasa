# Scrape competitor's data

- get page content using `requests`.
- find the table of content by finding 'table' Tag using `BeautifulSoup`.
- get each row in the table using `findAll` searching for 'tr' Tag.
- iterate over the rows and get all the cells by `findAll` searching for 'td' Tag.
- save the table content in a dataframe.

```python
import requests 
from bs4 import BeautifulSoup as bs
import pandas as pd
import numpy as np

url = 'https://www.spaceweatherlive.com/en/solar-activity/top-50-solar-flares'
response = requests.get(url)
page_soup = bs(response.text)
table_content = page_soup.find('table')
table_rows = table_content.tbody.findAll('tr')
rank = []
x_class = []
date = []
region = []
start_time =[]
max_time = []
end_time = []
movie = []
for row in table_rows:
    cells = row.findAll('td')
    rank.append(cells[0].text)
    x_class.append(cells[1].text)
    date.append(cells[2].text)
    region.append(cells[3].text)
    start_time.append(cells[4].text)
    max_time.append(cells[5].text)
    end_time.append(cells[6].text)
    movie.append(cells[7].text)

dFrame = pd.DataFrame({'Rank':rank, 'X_class':x_class, 'Date':date, 'Region':region, 'Start_time':start_time, 'Max_time':max_time, 'End_time':end_time, 'Movie':movie})
dFrame
```

# Scrape the NASA data

- get page content using `requests`.
- get the desired table using `BeautifulSoup` by finding 'pre' Tag.
- dealing with content as a Text, so split the text on `\n` to get it as list of lines.
- get ride of the unwanted lines inside the 'pre' Tag to isolate the rows of the table.
- then iterate over the lines and split the on the `space` to get each values and append it to lists.
>**Note:** Inner while loop is used to remove the unwanted valuse.
- creat a dataframe `nasaDataFrame`.

```python
nasaUrl = 'https://cdaw.gsfc.nasa.gov/CME_list/radio/waves_type2.html'
response_nasa = requests.get(nasaUrl)
```

```python
nasa_soup = bs(response_nasa.text)
string_content=nasa_soup.find('pre').text
List_string = string_content.split('\n')
List_string = List_string[12:-1]

startDate=[]
startTime = []
endDate = []
endTime = []
startFreq = []
endFreq = []
flareLoc = []
flareReg = []
importance = []
CME_Date = []
CME_Time = []
CPA = []
width = []
speed = []
plots = []
for row in List_string:
    row = row.split(' ')
    while '' in row:
        row.remove('')
    if len(row) > 12:
        startDate.append(row[0])
        startTime.append(row[1])
        endDate.append(row[2])
        endTime.append(row[3])
        startFreq.append(row[4])
        endFreq.append(row[5])
        flareLoc.append(row[6])
        flareReg.append(row[7])
        importance.append(row[8])
        CME_Date.append(row[9])
        CME_Time.append(row[10])
        CPA.append(row[11])
        width.append(row[12])
        speed.append(row[13])
        plots.append(row[14])
#the last row was messy so fix it manually.  
plots[-1] = speed[-1]
speed[-1] = width[-1]
width[-1] = CPA[-1]
CPA[-1] = CME_Time[-1]
CME_Time[-1] = CME_Date[-1]
CME_Date[-1] = importance[-1]
flareReg[-1] = '-----'
importance[-1] = '----'

nasaDataFrame = pd.DataFrame({'Start_Date':startDate,'Start_Time':startTime,'End_Date':endDate,'End_Time':endTime, 'Start_Frequency':startFreq, 'End_Frequency':endFreq,'Flare_Location':flareLoc, 'Flare_Region':flareReg,'importance':importance,'CME_Date':CME_Date,'CME_Time':CME_Time, 'CPA':CPA, 'Width':width,'Speed':speed, 'Plots':plots })
nasaDataFrame
```

# Tidy the NASA table


- as instructed in https://cdaw.gsfc.nasa.gov/CME_list/radio/waves_type2_description.htm, and looking at the data, knowing how the missing values was encoded.
- First replacing all the missing values with NaN.
- Then creat a column to identifie the `Halo CPA`.
>**Note:** if the value of CPA is Halo then its value in the column is `True` otherwise its value is `False`.
- creat a column to identifie the `lower bound width`, the number has `>` sign so is value is lower bound.
- also cleaning the column width, scince there is a value has character 'h'.
- at last merge column `Date` with Times columns.

```python
#Recode any missing entries as NaN
nasaDataFrame['Start_Frequency'] = nasaDataFrame['Start_Frequency'].replace('????','NaN')
nasaDataFrame['End_Frequency'] = nasaDataFrame['End_Frequency'].replace('????','NaN')
nasaDataFrame['Flare_Location'] = nasaDataFrame['Flare_Location'].replace('------','NaN')
nasaDataFrame['Flare_Region'] = nasaDataFrame['Flare_Region'].replace('-----','NaN')
nasaDataFrame['importance'] = nasaDataFrame['importance'].replace('----','NaN')
nasaDataFrame['CME_Date'] = nasaDataFrame['CME_Date'].replace('--/--' ,'NaN')
nasaDataFrame['CME_Time'] = nasaDataFrame['CME_Time'].replace('--:--','NaN')
nasaDataFrame['CPA'] = nasaDataFrame['CPA'].replace('----','NaN')
nasaDataFrame['Width'] = nasaDataFrame['Width'].replace('---','NaN')
nasaDataFrame['Width'] = nasaDataFrame['Width'].replace('----','NaN')
nasaDataFrame['Speed'] = nasaDataFrame['Speed'].replace('---','NaN')
nasaDataFrame['Speed'] = nasaDataFrame['Speed'].replace('----','NaN')

#create list of Is_Halo 
is_halo = []
for cell in nasaDataFrame['CPA']:
    if cell == 'Halo':
        is_halo.append(True)
    else:
        is_halo.append(False)
nasaDataFrame['CPA'] = nasaDataFrame['CPA'].replace('Halo','NA')
nasaDataFrame['Is_Halo'] = is_halo

width_lower = []
for i in range(len(nasaDataFrame)):
    if nasaDataFrame['Width'][i][0] == '>':
        width_lower.append('LowerBound')
        nasaDataFrame.at[i,'Width'] = nasaDataFrame['Width'][i].replace('>','')
    else:
        width_lower.append('')
    if nasaDataFrame['Width'][i][-1] == 'h':
        nasaDataFrame.at[i,'Width'] = nasaDataFrame['Width'][i].replace('h','')

#Add the new column to the dataframe.
nasaDataFrame['Width_lowerBound'] = width_lower

newNasaDF = nasaDataFrame.drop(columns=['Start_Date', 'End_Date', 'CME_Date'])

newNasaDF['Start_Time'] = nasaDataFrame['Start_Date']+" "+nasaDataFrame['Start_Time']
newNasaDF['End_Time'] = nasaDataFrame['End_Date']+" "+nasaDataFrame['End_Time']
newNasaDF['CME_Time'] = nasaDataFrame['CME_Date']+" "+nasaDataFrame['CME_Time']
newNasaDF
```

# Integration


To define a Matching function between the two datasets, we need to know the commen effective attributes like [`start time`, `importance`, `Regoin`]

### Steps
1- sort the nasa dataset to isolate the `X_class` importance, and process it (less computation).

2- get the indices of rows in `nasa` dataset that match the `Top50`, calculate the similarity between attributes using `SequenceMatcher` function in `difflib`, the match occurs if the sum of similarity > threshold.

3- after getting the indices get the matched rows from nasa dataset and save them in a new dataframe `matched_df`.

4- add `Rank` column in nasa dataframe, and return the `matched_df`.


>**Note:** Found 37 matches between the two datasets, the unmatched rows don't exist in nasa dataset. 

```python
from difflib import SequenceMatcher

# start time, importance, Regoin
def matching_function(nasaDF , Top50DF):
    #sort nasaDF to isolate the X class and NaN importance.
    sortedNasaDF = nasaDF.sort_values('importance', axis = 0, ascending = False, 
                 inplace = False, na_position ='last')
    
    sortedNasaDF = sortedNasaDF[:93]
    threshold = 2.1 #trying multiples values to get an accurate matching
    indices_holder = []
    for indexI,rowI in Top50DF.iterrows():
        Mrow = None
        BMax = -1
        BMindex=-1
        for indexJ, rowJ in sortedNasaDF.iterrows():
            similarity = SequenceMatcher(None, rowJ['Start_Time'],rowI['Start_time']).ratio()
            similarity += SequenceMatcher(None, rowJ['importance'],rowI['X_class']).ratio()
            similarity += SequenceMatcher(None, rowJ['Flare_Region'],rowI['Region']).ratio()
            if similarity > BMax and similarity > threshold and indexJ not in indices_holder:
                BMindex = indexJ
                BMax = similarity
        indices_holder.append(BMindex)
    matched_df = pd.DataFrame(sortedNasaDF, columns=list(sortedNasaDF.columns), index=indices_holder)
    #add Rank column to nasa dataframe
    if 'Rank' not in nasaDF.columns:
        nasaDF.insert(0,'Rank',['NaN' for i in range(len(nasaDF))])
    rank=1
    for val in indices_holder:
        if val!=-1:
            nasaDF.at[val,'Rank']=rank
        rank += 1
    return matched_df
    
    
    
# newNasaDF : the nasa data after the tyding step
# NDataFrame : the Top 50 solar flares in the SpaceWeatherLive data
best_matched = matching_function(newNasaDF, NDataFrame)
best_matched   
```
